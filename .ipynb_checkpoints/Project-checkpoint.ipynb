similar book prediction based on users feedback

```python
filtered_ratings=books_with_ratings[books_with_ratings["User-ID"].isin(power_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count().sort_values('Book-Rating',ascending=False)
```

```python
y
```

```python
 y.sort_values('User-ID',ascending=False)
```

```python
y=y['User-ID']>=50
y
```

```python
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt=pt.fillna(0)
```

```python
pt
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
type(similarity_scores)
```

```python
df_temp=pd.DataFrame(similarity_scores)
```

```python
df_temp
```

```python
pt.index
```

```python
df_temp_name=df_temp
```

```python
df_temp_name.index=pt.index
df_temp_name.columns=pt.index
```

```python
df_temp_name
```

```python
pt.index
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x : x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```
